### **ğŸ¯ í•œ ì¤„ ìš”ì•½**

````python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html>
<body>
    <div>
        <h1>í—¤ë”1</h1>
        <p>í—¤ë”1 ì— í¬í•¨ëœ ë³¸ë¬¸</p>
        <div>
            <h2>í—¤ë”2-1 ì œëª©</h2>
            <p>í—¤ë”2-1 ì— í¬í•¨ëœ ë³¸ë¬¸</p>
            <h3>í—¤ë”3-1 ì œëª©</h3>
            <p>í—¤ë”3-1 ì— í¬í•¨ëœ ë³¸ë¬¸</p>
            <h3>í—¤ë”3-2 ì œëª©</h3>
            <p>í—¤ë”3-2 ì— í¬í•¨ëœ ë³¸ë¬¸</p>
        </div>
        <div>
            <h2>í—¤ë”2-2 ì œëª©2</h2>
            <p>í—¤ë”2-2 ì— í¬í•¨ëœ ë³¸ë¬¸</p>
        </div>
        <br>
        <p>ë§ˆì§€ë§‰ ë‚´ìš©</p>
    </div>
</body>
</html>
"""

headers_to_split_on = [
    ("h1", "Header 1"),  # ë¶„í• í•  í—¤ë” íƒœê·¸ì™€ í•´ë‹¹ í—¤ë”ì˜ ì´ë¦„ì„ ì§€ì •í•¨
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]

# ì§€ì •ëœ í—¤ë”ë¥¼ ê¸°ì¤€ìœ¼ë¡œ HTML í…ìŠ¤íŠ¸ë¥¼ ë¶„í• í•˜ëŠ” HTMLHeaderTextSplitter ê°ì²´ë¥¼ ìƒì„±í•¨
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
# HTML ë¬¸ìì—´ì„ ë¶„í• í•˜ì—¬ ê²°ê³¼ë¥¼ html_header_splits ë³€ìˆ˜ì— ì €ì¥í•¨
html_header_splits = html_splitter.split_text(html_string)
# ë¶„í• ëœ ê²°ê³¼ë¥¼ ì¶œë ¥í•¨
for header in html_header_splits:
    print(f"{header.page_content}")
    print(f"{header.metadata}", end="\n=====================\n")
````

## ë‹¤ë¥¸ splitterì™€ íŒŒì´í”„ë¼ì¸ìœ¼ë¡œ ì—°ê²°í•˜ê³ , ì›¹ URLì—ì„œ HTMLì„ ë¡œë“œí•˜ëŠ” ê²½ìš°

ì´ ì˜ˆì‹œì—ì„œëŠ” ì›¹ URLë¡œë¶€í„° HTML ì½˜í…ì¸ ë¥¼ ë¡œë“œí•œ í›„, ì´ë¥¼ ë‹¤ë¥¸ splitterì™€ íŒŒì´í”„ë¼ì¸ìœ¼ë¡œ ì—°ê²°í•˜ì—¬ ì²˜ë¦¬í•˜ëŠ” ê³¼ì •


````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

url = "https://plato.stanford.edu/entries/goedel/"  # ë¶„í• í•  í…ìŠ¤íŠ¸ì˜ URLì„ ì§€ì •í•¨

headers_to_split_on = [  # ë¶„í• í•  HTML í—¤ë” íƒœê·¸ì™€ í•´ë‹¹ í—¤ë”ì˜ ì´ë¦„ì„ ì§€ì •í•¨
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4"),
]

# HTML í—¤ë”ë¥¼ ê¸°ì¤€ìœ¼ë¡œ í…ìŠ¤íŠ¸ë¥¼ ë¶„í• í•˜ëŠ” HTMLHeaderTextSplitter ê°ì²´ë¥¼ ìƒì„±í•¨
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)

# URLì—ì„œ í…ìŠ¤íŠ¸ë¥¼ ê°€ì ¸ì™€ HTML í—¤ë”ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶„í• í•¨
html_header_splits = html_splitter.split_text_from_url(url)

chunk_size = 500  # í…ìŠ¤íŠ¸ë¥¼ ë¶„í• í•  ì²­í¬ì˜ í¬ê¸°ë¥¼ ì§€ì •í•¨
chunk_overlap = 30  # ë¶„í• ëœ ì²­í¬ ê°„ì˜ ì¤‘ë³µë˜ëŠ” ë¬¸ì ìˆ˜ë¥¼ ì§€ì •í•¨
text_splitter = RecursiveCharacterTextSplitter(  # í…ìŠ¤íŠ¸ë¥¼ ì¬ê·€ì ìœ¼ë¡œ ë¶„í• í•˜ëŠ” RecursiveCharacterTextSplitter ê°ì²´ë¥¼ ìƒì„±í•¨
    chunk_size=chunk_size, chunk_overlap=chunk_overlap
)

# HTML í—¤ë”ë¡œ ë¶„í• ëœ í…ìŠ¤íŠ¸ë¥¼ ë‹¤ì‹œ ì²­í¬ í¬ê¸°ì— ë§ê²Œ ë¶„í• í•¨
splits = text_splitter.split_documents(html_header_splits)

# ë¶„í• ëœ í…ìŠ¤íŠ¸ ì¤‘ 80ë²ˆì§¸ë¶€í„° 85ë²ˆì§¸ê¹Œì§€ì˜ ì²­í¬ë¥¼ ì¶œë ¥í•¨
for header in splits[80:85]:
    print(f"{header.page_content}")
    print(f"{header.metadata}", end="\n=====================\n")
````

ë”°ë¼ì„œ "h1" ìš”ì†Œì™€ ê´€ë ¨ í…ìŠ¤íŠ¸ëŠ” ì²­í¬ ë©”íƒ€ë°ì´í„°ì— ë‚˜íƒ€ë‚˜ì§€ ì•Šì§€ë§Œ, í•´ë‹¹ë˜ëŠ” ê²½ìš° "h2"ì™€ ê´€ë ¨ í…ìŠ¤íŠ¸ëŠ” ë³¼ ìˆ˜ ìˆìŒ
